# Preparing the data



# First sugestion: TD-IDF - 
The tfâ€“idf value increases proportionally to the number of times a word appears in the document and is offset by the number of documents in the corpus that contain the word, which helps to adjust for the fact that some words appear more frequently in general. In other words, which words are proeminent in a certain document compared to other documents. 

Here we'll compare one document to all the others documents. We'll see latter a second suggestion where we find the topic of the document without comparing to other documents. 


```python
#opening the questions#
df = pd.read_excel('new_frame.xlsx')
df.head()

unique_documents = df['id'].unique() #a list with all documents in the csv. 
df = df.set_index('id') #set the document number as index of the dataframe

list_questions_per_document = []
for i in unique_documents:
    all_questions_per_document = df.loc[i,'questions_lemmatized'] #it localize all questions from the same document
    total_questions = '' #this is initial string that aggregates all questions
    try:
        for question in all_questions_per_document:
            question = re.sub(r'\W+', ' ', question).lower().replace('_', ' ')
            total_questions += (question + ' ') #we include the sample to avoid concatenation of words
        list_questions_per_document.append(total_questions) #this is the final document. Each group of questions per document is a string
    except: 
        continue
        
x = list_questions_per_document[0] #sample to test the similarity

list_questions_per_document = list_questions_per_document[1:]
```

```python
list_tokens_document = [] #we need to make a token. 
for i in list_questions_per_document:
        words = word_tokenize(i)
        final_words = []
        for word in words:
            if word not in stopwords.words('dutch'):
                if len(word) > 2:
                    final_words.append(word)
        list_tokens_document.append(final_words)



dictionary = Dictionary(list_tokens_document) #this creates the dictionary of words that will make the corpus
corpus = [dictionary.doc2bow(tokens_document) for tokens_document in list_tokens_document] #corpus

list_of_key_words = []
for i in range(len(corpus)):
    tfidf = TfidfModel(corpus)#generates the model
    doc = corpus[i] #it takes the key words for every document
    tfidf_weights = tfidf[doc] #pick the tfidf_weights for each set of 
    sorted_tfidf_weights = sorted(tfidf_weights, key=lambda w: w[1], reverse=True) #sort by value
    key_words = []
    for term_id, weight in sorted_tfidf_weights[:5]:
        print(str(dictionary.get(term_id)), weight)
        key_words.append(str(dictionary.get(term_id)))
    print('\n')
    list_of_key_words.append(key_words)
 
```

# Way to compare document similarity


```python

sims = gensim.similarities.Similarity(None, tfidf[corpus],
                                      num_features=len(dictionary))

query_doc = [w.lower() for w in word_tokenize(x)]
query_doc_bow = dictionary.doc2bow(query_doc)
query_doc_tf_idf =  tfidf[query_doc_bow]
sims = sims[query_doc_tf_idf]

sims = sorted(enumerate(sims), key=lambda item: -item[1])

print(x)
print('\n')

for i,w in sims[:5]:
    print(list_questions_per_document[i])
    print('\n')
```

# Second suggestion: LDA Topics

This approach is about mathematically finding the main topic in a certain document. Here, there is no comparison to other documents. 


```python
 #set the document number as index of the dataframe

# a list of string. Each string is the aggregation of all questions per document
list_of_questions_lda = []
for i in unique_documents:
    try:
        x = df.loc[i,'questions_lemmatized'] #it localize all questions from the same document
        total_questions = [] #this is initial string that aggregates all questions
        for q in x:
            q = re.sub(r'\W+', ' ', q).lower()
            total_questions.append(q)
        list_of_questions_lda.append(total_questions)
    except:
        continue



stop = set(stopwords.words('dutch'))
exclude = set(string.punctuation)
def clean(doc):
    stop_free = " ".join([i for i in doc.lower().split() if i not in stop if len(i) > 2])
    return stop_free       

for questions_per_doc in list_of_questions_lda:
    doc_clean = [clean(doc).split() for doc in questions_per_doc]
    dictionary = corpora.Dictionary(doc_clean)
    doc_term_matrix = [dictionary.doc2bow(doc) for doc in doc_clean]
    Lda = gensim.models.ldamodel.LdaModel
    ldamodel = Lda(doc_term_matrix, num_topics=1, id2word = dictionary, passes=100)
    topics = ldamodel.show_topics(num_topics=1, num_words=5,formatted=False)
    topics_words = [(tp[0], [wd[0] for wd in tp[1]]) for tp in topics]
        #Below Code Prints Only Words 
    for topic, words in topics_words:
        for i in range(len(topics_words)):
            print(" ".join(words))   
            print('\n')
```
